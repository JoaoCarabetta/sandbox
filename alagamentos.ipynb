```python
m = folium.Map(location=[-22.857, -43.216])
tiles = [Babel(grid).geo_to_tile(lat=-22.857, lon=-43.216, area_km=10) for grid in ['s2', 'h3']]
for tile in tiles:
    folium.GeoJson(tile.geometry.geojson).add_to(m)
m
```
